```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats

# przygotowanie danych
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
#zADANIE 1 - Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. 
collection_100_element = np.sort(stats.norm.rvs(3, 1, 100)).round(2)
collection_10000_element = np.sort(stats.norm.rvs(3, 1, 10000)).round(2)
```

```python
x100 = np.linspace(-10, 10, 100)
collection_100_element_normal_distribution= stats.norm.pdf(x100, 0, 1)
x10000 = np.linspace(-10, 10, 10000)
collection_10000_element_normal_distribution = stats.norm.pdf(x10000, 0, 1)
plt.figure(figsize=(3, 2))
plt.plot(collection_100_element, collection_100_element_normal_distribution, label='zbiór 100 elementów o rozkładzie normalnym (odchylenie standardowe = 1, średnia = 3)')
plt.legend(loc=(1.05,0.9))
plt.show()
plt.figure(figsize=(3, 2))
plt.plot(collection_10000_element, collection_10000_element_normal_distribution, label='zbiór 10 000 elementów o rozkładzie normalnym (odchylenie standardowe = 1, średnia = 3)')
plt.legend(loc=(1.05,0.9))
plt.show()
#Im zbiór ma więcej liczb tym zbiór jest zbliżony do ideału odwzorowania rozkłądu normalnego  , czy zbiór 10000 lepiej to odwzorowuje.
```

```python
#wykres z 10000 elementów , lepiej odwzorowuje rozład normalny niz zbiór ze 100 elementami.
```

```python
srednia_rn100 = np.round(np.mean(collection_100_element),2)
srednia_rn10000 = np.round(np.mean(collection_10000_element),2)
print(f'średnia dla x=100 równe {srednia_rn100}')
print(f'średnia dla x=10000 równe {srednia_rn10000}')

#Im więcej elementów posiada zbiór, to średnia wynosi 3.
```

```python
mediana_rn100 = np.round(np.median(collection_100_element),2)
mediana_rn10000 = np.round(np.median(collection_10000_element),2)
print(f'mediana dla x=100 równa {mediana_rn100}')
print(f'mediana dla x=10000 równa {mediana_rn10000}')
#Im zbiór ma więcej elementów tym mediana minimalnie idzie w góre. Media dla 100 jest równa średniej 3.
```

```python
moda_rn100 = stats.mode(collection_100_element)
moda_rn10000 = stats.mode(collection_10000_element)
print(f'moda dla x=100 równa {moda_rn100}')
print(f'moda dla x=10000 równa {moda_rn10000}')

# Różnice między średnią a modą są tym mniejsze, im bardziej liczebny jest zbiór
```

```python
quantile_0_rn100 = np.quantile(collection_100_element, 0.0)
quantile_1_rn100 = np.round(np.quantile(collection_100_element, 0.25),2)
quantile_3_rn100 = np.round(np.quantile(collection_100_element, 0.75),2)
quantile_4_rn100 = np.quantile(collection_100_element, 1.0)

quantile_0_rn10000 = np.quantile(collection_10000_element, 0.0)
quantile_1_rn10000 = np.round(np.quantile(collection_10000_element, 0.25),2)
quantile_3_rn10000 = np.round(np.quantile(collection_10000_element, 0.75),2)
quantile_4_rn10000 = np.quantile(collection_10000_element, 1.0)
print(f'''Q0_rn100: {quantile_0_rn100}
Q1_rn100: {quantile_1_rn100}
Q2_rn100: {mediana_rn100}
Q3_rn100: {quantile_3_rn100}
Q4_rn100: {quantile_4_rn100}

Q0_rn10000: {quantile_0_rn10000}
Q1_rn10000: {quantile_1_rn10000}
Q2_rn10000: {mediana_rn10000}
Q3_rn10000: {quantile_3_rn10000}
Q4_rn10000: {quantile_4_rn10000}''')
#Kwartyle Q1 i Q3 są do siebie  zbliżone w obu zbiorach.Kwartyle Q0 i Q4 różnią się ze względu na większą ilośc elementów zbioru 10000
```

```python
range_rn100 = max(collection_100_element) - min(collection_100_element)
range_rn10000 = max(collection_10000_element) - min(collection_10000_element)
print(f'''zakres zbioru 100-elementowego wynosi {range_rn100}
zakres zbioru 10 000-elementowego wynosi {range_rn10000}''')

#Wartości między zbiorami sa oddalone ze względu , ze drugi zbiór równy jest 10000 ,natomiast nie zaburza to zadanych wartości jak średnia=3 i odchylenie standardowe=1.
```

```python
iqr_rn100 = quantile_3_rn100 - quantile_1_rn100
iqr_rn10000 = quantile_3_rn10000 - quantile_1_rn10000
print(f'''rozstęp międzykwartylowy dla zbioru 100-elementowego wynosi {iqr_rn100}
rozstęp międzykwartylowy dla zbioru 10 000-elementowego wynosi {iqr_rn10000}''')

#Im większa liczebność zbioru, tym wartość IQR jest bardzo bliska wartości 1.34.
```

```python
plt.boxplot(collection_100_element, vert=False)
plt.show()
plt.boxplot(collection_10000_element, vert=False)
plt.show()

#większa liczebność zbioru sprawia , że jest więcej wartości odstających , natomiast nie zaburza to założeń zbioru ,średnia równa 3, odchylenie standardowe równe 1
```

```python
variance_rn100 = np.var(collection_100_element, ddof=1)
variance_rn10000 = np.var(collection_10000_element, ddof=1)
print(f'''wariancja dla zbioru 100-elementowego równe {variance_rn100}
wariancja dla zbioru 10 000-elementowego równe {variance_rn10000}''')
# Większa liczebność zbioru sprawia , że wariacja jest zbliożna do 1
```

```python
s_rn100 = variance_rn100**(1/2)
s_rn10000 = variance_rn10000**(1/2)
print(f'''odchylenie standardowe dla zbioru 100-elementowego równe {s_rn100}
odchylenie standardowe dla zbioru 10 000-elementowego równe {s_rn10000}''')
```

```python
standard_deviation_rn100 = np.std(collection_100_element, ddof=1)
standard_deviation_rn10000 = np.std(collection_10000_element, ddof=1)
print(f'''odchylenie standardowe dla zbioru 100-elementowego równe {standard_deviation_rn100}
odchylenie standardowe dla zbioru 10 000-elementowego równe {standard_deviation_rn10000}''')
# Większa liczebność zbioru sprawia , że średnia jest zbliożna do 1
```
